```python
# importing the packages
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

df = pd.read_csv(r'C:\Users\saisr\OneDrive\Bilder\Dokumente\Desktop\DATA Working with\movies.csv')
```

```python
df.head()
```

```python
# We need to see if we have any missing data
# Let's loop through the data and see if there is anything missing

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,round(pct_missing*100)))
```

```python
#dropping empty rows
df = df.dropna()
```

```python
print(df.dtypes)
```

```python
#changing datatype
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
#extracting year from released
df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
df.head()
```

```python
#sort by gross
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
df.head()
```

```python
# Looks at company
df.head()
```

```python
df_numerized = df
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype =='object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
df_numerized.head()
```

```python
correltion_matrix = df_numerized.corr(numeric_only=True, method = 'pearson')

sns.heatmap(correltion_matrix, annot = True)

plt.title('Correltion Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
df_numerized.corr()
```

```python
correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs.head(20)
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs.head(20)
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```
